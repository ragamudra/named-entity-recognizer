Next section deals with word vectors. Way to represent a word into set of numbers or vectors. These vectors can then be used as features in model

Explaining vector values of different words in Spacy The output in the following is a vector representation of the word “This”. Spacy has trained a word2vec model to arrive at 300 long dimensions for each word

```python
doc = nlp("This is to explain parts of speech in Spacy")
for i in doc:
    print(i.text, i.vector)
```

We will take the example of the words - “learn”,”read”, “understand”. In order to contrast the similarities we will take the context vector of the word “run”

```python
import spacy
nlp = spacy.load("en_core_web_md")
doc1 = nlp(u'learn')
doc2 = nlp(u'read')
doc3 = nlp(u'understand')
doc4 = nlp(u'run')
for i in [doc1,doc2,doc3,doc4]:
    for j in [doc1,doc2,doc3,doc4]:
        
        if((i.vector_norm>0) & (j.vector_norm>0)): 
            print (i,j,i.has_vector,j.has_vector,i.similarity(j))
```
