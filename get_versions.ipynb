```python
import pkg_resources
import pandas as pd
installed_packages = pkg_resources.working_set
installed_packages_list = sorted(["%s==%s" % (i.key, i.version)
   for i in installed_packages])
df1 = pd.DataFrame(installed_packages_list)
df1.columns = ["pack"]
df1["pack_name"] = df1["pack"].str.split("==").str.get(0)
df1["version"] = df1["pack"].str.split("==").str.get(1)
```

```python
df1.to_clipboard()
```
